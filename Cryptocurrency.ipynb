```python
import yfinance as yf
data = yf.download("BTC-USD", start="2022-01-01", end="2023-01-01")
btc_close = data['Close']
```

```python
import matplotlib.pyplot as plt
btc_close.plot(title="Bitcoin Closing Price")
plt.show()
```

```python
btc_diff = btc_close.diff().dropna()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(btc_close, order=(5,1,0))  
model_fit = model.fit()
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=10)
print("Next 10-day forecast:")
print(forecast)

# Plot
btc_close.plot(label='Historical')
forecast.plot(label='Forecast', legend=True, title="Bitcoin Forecast")
plt.show()
```
